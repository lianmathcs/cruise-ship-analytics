# Exploratory Data Analysis (EDA) â€“ Global Cruise Ship Dataset
- This notebook analyzes a cleaned global cruise ship dataset to uncover patterns in ship size, capacity, age, ship type, and operational characteristics, enabling comparisons across cruise fleets and highlighting structural trends in the global cruise industry
- The EDA focuses on distributions, correlations, group-level comparisons, and statistical testing (including correlation analysis and regression-based tests) to support clear, data-driven insights

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import ttest_ind
from scipy import stats
```

```python
df = pd.read_csv("../data/cruise_ships_eda.csv")
# we will look at the general columns information 
df.info()
```

```python
df.describe().T
```

### Data Quality Note ###
- Some ships had implausible ages (e.g., 9000 years), which correspond to planned or future builds in the source website. 
These values were treated as invalid and set to missing to avoid skewing the analysis.
- Crew size contained extreme outliers due to scraping errors. I used domain knowledge to cap impossible values and retained realistic observations for analysis.

```python
# Count outliers before cleaning
print("DATA CLEANING SUMMARY")
print("="*60)
print(f"Ships with crew > 2500: {(df['crew'] > 3000).sum()} rows")
print(f"Ships with crew < 5: {(df['crew'] < 5).sum()} rows")
print(f"Ships with age > 100: {(df['age'] > 100).sum()} rows")

# Clean outliers directly in original columns
df.loc[df["crew"] > 2500, "crew"] = np.nan  # Extreme outliers
df.loc[df["crew"] < 5, "crew"] = np.nan     # Unrealistically small
df.loc[df["age"] > 100, "age"] = np.nan     # Implausible ages

print("\nCLEANING COMPLETE")
print("="*60)
print(f"Remaining ships with valid crew: {df['crew'].notna().sum()}")
print(f"Remaining ships with valid age: {df['age'].notna().sum()}")
```

```python
df.describe().T
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
# 1) Crew vs Passengers
axes[0].scatter(df['crew'], df['passengers_clean'], alpha=0.6)
axes[0].set_title('Crew Size vs Passenger Capacity')
axes[0].set_xlabel('Crew')
axes[0].set_ylabel('Passengers')
axes[0].grid(True, alpha=0.3)

# 2) Age vs Gross Tonnage
axes[1].scatter(df['age'], df['gross_tonnage'], alpha=0.6)
axes[1].set_title('Ship Age vs Gross Tonnage')
axes[1].set_xlabel('Age (years)')
axes[1].set_ylabel('Gross Tonnage')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### Crew Size vs Passenger Capacity
- The scatter plot shows a strong positive relationship between crew size and passenger capacity. As passenger capacity increases, crew size generally increases as well, indicating that staffing scales with ship size to support onboard operations and service levels.

- However, the relationship is not perfectly linear, with noticeable dispersion among larger ships. This suggests that while crew size is a key driver of passenger capacity, operational strategy, ship type, and service standards also influence staffing requirements.


### Ship Age vs Gross Tonnage
- The relationship between ship age and gross tonnage appears weak and non-linear. Newer ships tend to span a wide range of gross tonnage values, while older ships are generally smaller, though with notable exceptions.

- This pattern suggests that ship size is driven more by design and market positioning than by age alone. While recent decades have seen the introduction of very large vessels, ship age by itself does not strongly predict gross tonnage.

### Key takeaway
- Larger ships usually require more crew, but staffing intensity varies across ships and ship types, reflecting differences in service models and cruise line practices.

- Gross tonnage is influenced primarily by ship design and cruise line strategy rather than vessel age.

- Initial EDA established core structural relationships between ship size, capacity, and labor, which motivated the creation of normalized efficiency metrics to enable following fair comparisons across ship generations and categories 

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# 1) Length vs Passengers
axes[0].scatter(df['length_m'], df['passengers_clean'], alpha=0.6)
axes[0].set_title('Ship Length vs Passenger Capacity')
axes[0].set_xlabel('Length (m)')
axes[0].set_ylabel('Passengers')
axes[0].grid(True, alpha=0.3)

# 2) Gross Tonnage vs Passengers
axes[1].scatter(df['gross_tonnage'], df['passengers_clean'], alpha=0.6)
axes[1].set_title('Gross Tonnage vs Passenger Capacity')
axes[1].set_xlabel('Gross Tonnage')
axes[1].set_ylabel('Passengers')
axes[1].grid(True, alpha=0.3)

# 3) Beam vs Length
axes[2].scatter(df['beam_m'], df['length_m'], alpha=0.6)
axes[2].set_title('Beam vs Length')
axes[2].set_xlabel('Beam (m)')
axes[2].set_ylabel('Length (m)')
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### Ship Dimensions and Passenger Capacity  
   - Across the three visualizations, passenger capacity is shown to scale primarily with overall ship size and volume rather than any single physical dimension. Both ship length and gross tonnage exhibit positive relationships with passenger capacity, with gross tonnage displaying the strongest and most consistent association, reflecting its role as a comprehensive measure of internal volume.

- The relationship between beam and length further illustrates that cruise ship dimensions scale proportionally, following standardized design and engineering constraints. This proportional scaling explains why length alone does not fully capture passenger capacity, as ships of similar length can differ in width, layout, and internal volume.

### Overall takeaway
- Together, these plots show how physical dimensions set the limits on passenger capacity and ship design. While length, beam, and gross tonnage all increase as ships get larger, gross tonnage has the strongest relationship with passenger capacity, highlighting the importance of usable interior volume. The variation seen across ships with similar dimensions points to differences in design choices, which motivates the use of normalized metrics to compare space and design efficiency more fairly across vessels

```python
# Distribution of Key Ship Characteristics
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

metrics = ['crew', 'age', 'passengers_clean', 'gross_tonnage']
titles = ['Crew Size', 'Ship Age', 'Passenger Capacity', 'Gross Tonnage']

for idx, (metric, title) in enumerate(zip(metrics, titles)):
    axes[idx].boxplot(df[metric].dropna())
    axes[idx].set_title(title)
    axes[idx].set_ylabel(metric)

plt.tight_layout()
plt.show()
```

### Distribution of Key Ship Characteristics

- The boxplots summarize the distributions of crew size, ship age, passenger capacity, and gross tonnage, highlighting both central tendencies and the presence of outliers across the global cruise fleet.

- All four variables exhibit right-skewed distributions, with a relatively small number of ships at the high end of size and capacity. In particular, passenger capacity and gross tonnage show substantial dispersion and extreme upper outliers, reflecting the existence of a limited number of mega-ships that are significantly larger than the rest of the fleet.

- Ship age displays a wider spread, indicating a mix of older vessels and newer builds currently in operation. Crew size follows a similar pattern to passenger capacity, suggesting that staffing levels scale with ship size while still varying across ship types and service models.

### Overall takeaway
- The cruise fleet is highly heterogeneous, with most ships clustered at smaller to mid-range sizes and a small number of very large vessels driving the upper extremes. This skewness and presence of outliers should be considered in downstream analysis and modeling, including the potential use of transformations or robust metrics.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Ships built per year
df['year_built'].value_counts().sort_index().plot(ax=axes[0])
axes[0].set_title('Ships Built Per Year')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Number of Ships')
axes[0].grid(True, alpha=0.3)

# Average size over time
df.groupby('year_built')['gross_tonnage'].mean().plot(ax=axes[1], color='orange')
axes[1].set_title('Average Gross Tonnage Over Time')
axes[1].set_xlabel('Year Built')
axes[1].set_ylabel('Avg Gross Tonnage')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### Shipbuilding Trends Over Time
- The â€œShips Built Per Yearâ€ plot shows clear variation in cruise ship construction activity over time. Shipbuilding is not evenly distributed across years, with noticeable periods of higher production followed by slower intervals. These fluctuations likely reflect broader economic cycles, changes in travel demand, and shifts in cruise industry investment.
- The â€œAverage Gross Tonnage Over Timeâ€ plot highlights a strong long-term trend toward larger ships. While there is year-to-year variability, the overall pattern shows that ships built in more recent years tend to have substantially higher average gross tonnage than those built earlier. This suggests a steady move toward larger, more complex vessels with greater onboard capacity and amenities.
### Overall takeaway
- Together, these plots show that the cruise industry has not only changed in how many ships it builds over time, but also in the type of ships it builds. Construction activity varies by period, but average ship size has generally increased, reinforcing earlier observations that modern ships are significantly larger than older vessels. This historical shift helps explain patterns observed in capacity, staffing, and cost metrics and supports the use of age-aware and size-normalized comparisons in later analysis

```python
# Just the 4 most important metrics
key_vars = ['crew', 'age', 'passengers_clean', 'gross_tonnage']

sns.pairplot(df[key_vars].dropna(), 
             height=2,
             plot_kws={'alpha': 0.6, 's': 20})
plt.suptitle('Key Ship Metrics Relationships', y=1.0)
plt.show()
```

### Overtakeaway
- The pairplot summarizes how ship size, capacity, and staffing scale together, while showing that ship age has a looser and less direct relationship with these core metrics

```python
EUR_TO_USD = 1.17

df["building_cost_usd_million"] = df.apply(
    lambda row: row["building_cost_million"] * EUR_TO_USD
    if row["building_cost_currency"] == "EUR"
    else row["building_cost_million"],
    axis=1
)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].scatter(
    df["building_cost_usd_million"],
    df["gross_tonnage"],
    alpha=0.6
)
axes[0].set_title("Building Cost vs Gross Tonnage (USD)")
axes[0].set_xlabel("Cost (Million USD)")
axes[0].set_ylabel("Gross Tonnage")
axes[0].grid(True, alpha=0.3)

axes[1].scatter(
    df["building_cost_usd_million"],
    df["passengers_clean"],
    alpha=0.6
)
axes[1].set_title("Building Cost vs Passenger Capacity (USD)")
axes[1].set_xlabel("Cost (Million USD)")
axes[1].set_ylabel("Passengers")
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### Overtakeaway
- After normalizing building costs to a common currency, cost shows a clear positive relationship with both gross tonnage and passenger capacity, indicating that capital investment scales with ship size, while variation reflects differences in design complexity and market positioning.

```python
# Calculate correlations
corr_matrix = df[['crew', 'age', 'passengers_clean', 'gross_tonnage', 
                  'length_m', 'speed_kn']].corr()
# Visualize
plt.figure(figsize=(8, 6))
plt.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar(label='Correlation')
plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=45)
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)
plt.title('Correlation Between Ship Metrics')

# Add correlation values
for i in range(len(corr_matrix.columns)):
    for j in range(len(corr_matrix.columns)):
        plt.text(j, i, f'{corr_matrix.iloc[i, j]:.2f}', 
                ha='center', va='center', color='white' if abs(corr_matrix.iloc[i, j]) > 0.5 else 'black')

plt.tight_layout()
plt.show()
print("KEY INSIGHTS:")
print("="*75)

# Q1: Crew ratios
median_ratio = df['crew_per_passenger'].median()
print(f"1. Median crew-to-passenger ratio: 1:{1/median_ratio:.1f} (or {median_ratio:.3f} crew per passenger)")

# Q2: Luxury indicator (space per passenger)
median_space = df['tonnage_per_passenger'].median()
print(f"2. Median space per passenger: {median_space:.1f} gross tons")

# Q3: Speed vs size correlation
speed_size_corr = df['speed_kn'].corr(df['gross_tonnage'])
print(f"3. Correlation between speed and size: {speed_size_corr:.3f}")

# Q4: Top performers
print("\n4. Top 5 ships by space per passenger (luxury):")
luxury_ships = df.nlargest(5, 'tonnage_per_passenger')[['name', 'tonnage_per_passenger', 'passengers_clean']]
print(luxury_ships.to_string(index=False))

print("\n5. Most efficient (passengers per crew):")
df['passengers_per_crew'] = df['passengers_clean'] / df['crew']
efficient_ships = df.nlargest(5, 'passengers_per_crew')[['name', 'passengers_per_crew', 'crew', 'passengers_clean']]
print(efficient_ships.to_string(index=False))
```

### Key Insights from Exploratory Analysis
- The median crew-to-passenger ratio across the dataset is approximately 1:2.9, or 0.35 crew members per passenger.
This suggests that, for a typical vessel, staffing scales closely with passenger volume, but not on a one-to-one basis. The ratio reflects a balance between operational needs and efficiency, with notable variation across ship types and service models.
- The median space allocation is 27.8 gross tons per passenger, providing a useful benchmark for comparing vessel density. Ships with higher values tend to emphasize comfort, amenities, or specialized operations, while lower values indicate higher-density, mass-market designs
- The correlation between ship speed and size (gross tonnage) is 0.253, indicating a weak positive relationship.
This suggests that larger ships are slightly more likely to be faster, but speed is primarily driven by design choices and operational intent rather than size alone. In other words, bigger ships are not necessarily much faster.

- !!!The ships with the highest space-per-passenger values are icebreakers, while those with the highest passengers-per-crew ratios are ferries. These vessels serve fundamentally different purposes than cruise and river cruise ships, which explains their extreme metric values. As a result, global rankings based on these metrics primarily reflect vessel function rather than passenger experience or service level.


```python

def extract_ship_type(name):
    name = str(name).lower()
    if "ferry" in name:
        return "Ferry"
    elif "icebreaker" in name:
        return "Icebreaker"
    elif "barge" in name:
        return "River Cruise"
    elif "river" in name:
        return "River Cruise"
    elif "cruise" in name:
        return "Cruise"
    else:
        return "Cruise"
df["ship_type"] = df["name"].apply(extract_ship_type)
df["ship_type"].value_counts()
```

### Data Quality Note
- Since ship type was not explicitly available, it was inferred from vessel names using keyword matching. Although this approach isnâ€™t exact, it effectively distinguishes between ships with very different purposes (for example, cruise ships versus ferries or icebreakers), which helps keep the exploratory analysis focused and interpretable.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 1) Crew per Passenger by Ship Type
sns.boxplot(
    data=df,
    x="ship_type",
    y="crew_per_passenger",
    ax=axes[0]
)
axes[0].set_title("Crew per Passenger Ratio by Ship Type")
axes[0].set_xlabel("Ship Type")
axes[0].set_ylabel("Crew per Passenger")
axes[0].tick_params(axis="x", rotation=45)

# 2) Tonnage per Passenger by Ship Type
sns.boxplot(
    data=df,
    x="ship_type",
    y="tonnage_per_passenger",
    ax=axes[1]
)
axes[1].set_title("Tonnage per Passenger by Ship Type")
axes[1].set_xlabel("Ship Type")
axes[1].set_ylabel("Tonnage per Passenger")
axes[1].tick_params(axis="x", rotation=45)
axes[0].set_yscale("log")
axes[1].set_yscale("log")

plt.tight_layout()
plt.show()
```

### Key Insights  
- River cruise ships and ferries exhibit more consistent crew-per-passenger ratios, with fewer extreme values, while ocean cruise ships and icebreakers show greater variability due to differences in size, service level, and operational purpose

- In the tonnage-per-passenger boxplot, river cruise ships show a tight distribution with few outliers, reflecting standardized vessel designs and similar passenger capacities. Icebreakers also appear to have few outliers, but for a different reason: their tonnage-per-passenger values are consistently extreme due to very low passenger counts and large hull sizes, causing most values to fall within the interquartile range. In contrast, ferries and ocean cruise ships display greater variability and more outliers, driven by diverse design choices and operational models

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df,
    x="passengers_clean",
    y="crew",
    hue="ship_type",
    alpha=0.7
)
plt.title("Passenger Capacity vs Crew by Ship Type")
plt.xlabel("Passenger Capacity")
plt.ylabel("Crew")
plt.legend(title="Ship Type")
plt.show()
```

### Key insights
- Crew size generally increases with passenger capacity, but the relationship differs clearly by ship type. Cruise ships show wide variation, river cruises are more standardized, ferries prioritize high passenger throughput with minimal crew, and icebreakers follow a distinct, non-passenger-focused staffing model.

```python
plt.figure(figsize=(10, 5))
sns.lineplot(
    data=df.sort_values("year_built"),
    x="year_built",
    y="crew_per_passenger",
    errorbar=None,
    marker="o",
    alpha=0.7
)
plt.title("Crew-to-Passenger Ratio Over Time")
plt.xlabel("Year Built")
plt.ylabel("Crew per Passenger")
plt.grid(True)
plt.show()
```
